```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
traindata = pd.read_csv("KNN_train.csv")
traindata = traindata.drop(["loan_id","proof_submitted"], axis = 1)
testdata = pd.read_csv("KNN_test.csv")
testdata = testdata.drop(["loan_id","proof_submitted"], axis = 1)
```

```python
X = traindata.iloc[:,:-1]
y = traindata.iloc[:,-1]
print("X: ", X)
print("y: ", y)
X["education"].fillna(0, inplace = True)
```

```python
#Finding the new column for the test data based on the model.
Xtest = testdata
Xtest["education"].fillna(0, inplace = True)
print("X: ", Xtest)
# y_predict=knn.predict(Xtest)
# print("Predicted y value for test data: \n", y_predict)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
ypred=knn.predict(Xtest)
print(ypred)
```
