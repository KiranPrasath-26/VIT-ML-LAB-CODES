```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing,svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,classification_report
```

```python
data=pd.read_csv("Iris.csv")
data = data.drop("Id", axis=1)
```

```python
data.head()
```

```python
X=data.iloc[:,0:-1]
X
y = data.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y=le.fit_transform(y)
print(y)
# X = X.apply(le.fit_transform)
# print(X)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)
print(xtrain,ytrain)
```

```python
model=svm.SVC()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
print(ypred)
print(ytest)
```

```python
print(model.score(xtest,ytest))
```

```python
Xtrain = train_data.iloc[:,1:]
ytrain = train_data.iloc[:,0]
print(Xtrain)
print(ytrain)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
# ytrain = le.fit_transform(ytrain)
# print(ytrain)
Xtrain["US"] = le.fit_transform(Xtrain["US"].astype(str))
Xtrain["Location"] = le.fit_transform(Xtrain["Location"].astype(str))
Xtrain["Urban"] = le.fit_transform(Xtrain["US"].astype(str))
print(Xtrain)
```

```python
Xtest = train_data.iloc[:,1:]
ytest = train_data.iloc[:,0]
# print(Xtest)
print(ytest)
Xtest["US"] = le.fit_transform(Xtest["US"].astype(str))
Xtest["Location"] = le.fit_transform(Xtest["Location"].astype(str))
Xtest["Urban"] = le.fit_transform(Xtest["US"].astype(str))
# print(Xtest)
```

```python
model=svm.SVR()
model.fit(Xtrain,ytrain)
ypred=model.predict(Xtest)
print("the predicted y values: ",ypred)
print("the test y values: ",ytest)
print("the score for the training data set is: ",model.score(Xtest,ytest))
# con_mat = confusion_matrix(ytest,ypred)
# print("Confusion matrix: \n", con_mat)
# print(classification_report(ytest, ypred))
```
