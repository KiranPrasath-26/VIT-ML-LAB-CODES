```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(1 , figsize = (15 , 7))
plt.title('Scatter plot of Age v/s Spending Score', fontsize = 20)
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.scatter( x = 'Age', y = 'Spending Score (1-100)', data = df, s = 100)
plt.show()
```

```python
X1 = df[['Age' , 'Spending Score (1-100)']].iloc[: , :].values
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(X1)  
print(y_predict)
```

```python
plt.scatter(X1[y_predict == 0, 0], X1[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
plt.scatter(X1[y_predict == 1, 0], X1[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
plt.scatter(X1[y_predict== 2, 0], X1[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
plt.scatter(X1[y_predict == 3, 0], X1[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster  
plt.scatter(X1[y_predict == 4, 0], X1[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') #for fifth cluster  
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')   
plt.title('Clusters of customers')  
plt.xlabel('Age')  
plt.ylabel('Spending Score (1-100)')  
plt.legend()  
plt.show()  
```
