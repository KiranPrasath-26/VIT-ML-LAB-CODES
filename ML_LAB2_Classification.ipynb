```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix, recall_score, precision_score
```

```python
data = pd.read_csv("play_tennis.csv")
data = data.drop(["day"],axis = 1)
data.head()
```

```python
X = data.iloc[:,:4]
y = data.iloc[:,-1]
print("independant variables: \n", X)
print("Dependant var: \n", y)
```

```python
lab_enc = preprocessing.LabelEncoder()
for i in range(4):
  X.iloc[:,i] = lab_enc.fit_transform(X.iloc[:,i])
y = lab_enc.fit_transform(y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.5, random_state=0)
print("Traindata: \n", xtrain, "\n", ytrain)
print("Testdata: \n", xtest, "\n", ytrain)
```

```python
gaus = GaussianNB()
mod = gaus.fit(xtrain,ytrain)
ypred = mod.predict(xtest)
print(ypred)
print(ytest)
print("MSE: ", mean_squared_error(ypred,ytest))

con_mat = confusion_matrix(ytest,ypred)
print("Confusion matrix: \n", con_mat)
```

```python
print(mod.score(xtest,ytest))
print(recall_score(ytest,ypred))
print(precision_score(ytest,ypred))
```

```python
#specificity
spec = con_mat[0,0]/con_mat[0,0]+con_mat[1,1]
#sensitivity
sens = con_mat[0,1]/con_mat[1,0]+con_mat[0,1]
print(spec)
print(sens)
```
