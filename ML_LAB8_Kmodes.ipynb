**KMODES**

```python
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Buy_Computer.csv')
df
```

```python
X = df.drop(['id', 'Buy_Computer'], axis=1)
```

```python
model = KModes(n_clusters=2,n_init=5)
model.fit(X)
```

```python
y_pred = model.predict(X)
```

```python
df['pred'] = y_pred
```
